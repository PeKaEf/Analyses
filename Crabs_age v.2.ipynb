```python
import matplotlib.pyplot as plt
import seaborn as sns

train_extended.drop(columns=("id"),inplace=True)

features=train_extended.columns.to_list()
features.remove('Age')
for feature in features:
    plt.scatter(x=feature,y='Age',data=train_extended,alpha=.1)
    plt.xlabel(feature)
    plt.ylabel('Age')
    plt.title(f'Crab age vs {feature}')
    plt.show()
```

Observations:
- we have one numeric value for 'sex' - to be replaced with mode of this feature
- females and males crabs don't occur in the groups of the youngest age; youngest crabs are of indeterminate sex, but at the same time they aren't the oldest ones
- we have two distant observations of length, which are outliers and should be deleted
- values over 2.0 of diameter look like outliers
- values over 1.0 of height look like outliers
- values over 20.0 of viscera weight look like outliers
- values over 35.0 of shell weight look like outliers
- height, diameter and length of values 0 should be deleted

Conclusions:
- it seems there is correlation between age and sex - crabs of age of 1, 2, 3 months are more likely to be of the indeterminate sex than of the male one, while crabs over 24 months are much more likely to be male or female
- the kind of 'turning point' is the age of 5 months, after which most of the features increases for most of the instances
- there are yet some instances, where the age > 5 month doesn't mean the increase of the features, this pertains also to length and height - we can see that these features don't incraese so much with the age growth
- the shell weight + the shucked weight doesn't give in total the weight and supposedly they should - maybe the reason is the water, as observed in the abalone dataset (https://shapbio.me/courses/biolB215s14/abalone_cleaning.html); we will calculate the weight of the shell by substracting the shucked weight from the total weight and compare it with the weight of the shell; if any instance will have greater shell weight than the calculated shell weight, it will be deleted

We shouldn't use the extended set for validation, so let's generate random numbers from range of 0-74052, which corresponds to the number of rows of the training set.
```python
te2=train_extended_.copy()
te2.reset_index(inplace=True)
te2
```

```python
X = te2.drop(["Age","index"],axis=1)
y = te2["Age"]

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error
from numpy import absolute
from math import sqrt

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, shuffle=True, random_state=42)

xgbreg=XGBRegressor()
random_forest_reg=RandomForestRegressor()
scaler=StandardScaler()
pipeline=Pipeline(steps=[("scaler",scaler),("random_forest_reg",random_forest_reg)])
rfc={'random_forest_reg__n_estimators':[100,200,300],'random_forest_reg__min_samples_split':[2,3,4,5],'random_forest_reg__bootstrap': [True, False]}
parameter_search=GridSearchCV(pipeline,param_grid=rfc,scoring='neg_mean_squared_error')
parameter_search.fit(X_train, y_train)
print("The best parameters are:", parameter_search.best_params_)
model=parameter_search.best_estimator_
y_predict_train=model.predict(X_train)
print(f'Train RMSE: {np.sqrt(mean_squared_error(y_train,y_predict_train))}')
y_predict_test=model.predict(X_test)
print(f'Test RMSE: {np.sqrt(mean_squared_error(y_test,y_predict_test))}')
```
