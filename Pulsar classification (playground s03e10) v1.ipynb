This is a binary classification task aimed at predicting whether an instance can be categorized as pulsar or not. All features are numerical and are related to selected measurements of each instance. The description of the features:
- Mean_Integrated: Mean of Observations
- SD: Standard deviation of Observations
- EK: Excess kurtosis of Observations
- Skewness: In probability theory and statistics, skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. Skewness of Observations.
- Mean _ DMSNR _ Curve: Mean of DM SNR CURVE of Observations (DM SNR Curve - Radio waves emitted from pulsars reach earth after traveling long distances in space which is filled with free electrons. The important point is that pulsars emit a wide range of frequencies, and the amount by which the electrons slow down the wave depends on the frequency. Waves with higher frequency are sowed down less as compared to waves with higher frequency. It means dispersion.)
- SD _ DMSNR _ Curve: Standard deviation of DM SNR CURVE of Observations
- EK _ DMSNR _ Curve: Excess kurtosis of DM SNR CURVE of Observations
- Skewness _ DMSNR _ Curve: Skewness of DM SNR CURVE of Observations

```python
features=train_df.columns.tolist()
features.remove("Class")
features.remove("id")
for feature in features:
    plt.figure(figsize=(8,8))
    sns.distplot(train_df.loc[train_df['Class']==0,feature],kde=True,color='orange',label='Non-pulsar')
    sns.distplot(train_df.loc[train_df['Class']==1,feature],kde=True,color='green',label='Pulsar')
    plt.legend(loc='upper right')
    plt.show()
```

It is shown above that all features are greatly correlated with the target values and should be kept for training. However, mostly they have far from normal distributions so they will require transformation in preprocessing.

```python
plt.figure(figsize=(9,9))
ax=sns.heatmap(train_df[features].corr(),xticklabels=train_df[features].corr().columns,yticklabels=train_df[features].corr().columns,cmap='RdYlGn',annot=True)
plt.show()
```

```python
from scipy import stats
corr={}
for f in features:
    corr[f]=stats.pointbiserialr(train_df[f],train_df['Class'])[0]
pd.Series(corr).sort_values(ascending=False)
```

As it can be seen, 'EK'-'Skewness' and 'EK_DMSNR_Curve'-'Skewness_DMSNR_Curve' are correlated to each other, but they are also highly correlated with the target, so we won't drop any of them.

```python
train_df.drop(columns=['id'],inplace=True)
test_df.drop(columns=['id'],inplace=True)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import f1_score
from sklearn.preprocessing import PowerTransformer, StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline

X=train_df.drop(['Class'],axis=1)
y=train_df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, shuffle=True, random_state=0)

ada=AdaBoostClassifier()
poly_features=PolynomialFeatures()
scaler=StandardScaler()

pipeline=Pipeline([("scaler",scaler),("poly_features",poly_features),("ada",ada)])

params={'poly_features__degree':[1,2,3],'ada__n_estimators':[50,100],'ada__learning_rate':[0.1, 0.3, 0.7, 1.0]}
gsearch1=GridSearchCV(pipeline,param_grid=params,scoring='f1_macro')
gsearch1.fit(X_train, y_train)
print("The best parameters are:", gsearch1.best_params_)
model2=gsearch1.best_estimator_
print(model2.score(X_train,y_train))
```

```python
print(model2.score(X_test,y_test))
```

```python
y_pred=model2.predict(X_test)
print(f1_score(y_test,y_pred))
```

Both the test and the train results are very good, so we will leave this model as version 1. Yet, we will try to improve it.
