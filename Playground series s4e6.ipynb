Basing on the above data, the following values of certain features can be removed without impacting greatly the training set:
- 'Application order' - values 0 and 9
- 'Curricular units 1st sem (enrolled)' - values over 21
- 'Curricular units 1st sem (evaluations)' - here all values will be kept, because the highest values are not the least frequent ones
- 'Curricular units 1st sem (approved)' - values over 20
- 'Curricular units 2nd sem (enrolled)' - values over 20
- 'Curricular units 2nd sem (approved)' - values over 19

'Course' feature has also some outliers and also its values are much higher than the values of other features. On second thoughts, it is decided to remove this feature from the training set, because it's not so prominent in distinguishing between the target values. We could rescale all the features, but as it looks like there are better features in the dataset, we will first remove 'Course' and if the model underfits, we will decide if this feature should be taken again into consideration.

```python
filter1=train_df[train_df['Application order']==0].index
train_df.drop(filter1,inplace=True)
filter2=train_df[train_df['Application order']==9].index
train_df.drop(filter2,inplace=True)
filter3=train_df[train_df['Curricular units 1st sem (enrolled)']>21].index
train_df.drop(filter3,inplace=True)
filter4=train_df[train_df['Curricular units 1st sem (approved)']>20].index
train_df.drop(filter4,inplace=True)
filter5=train_df[train_df['Curricular units 2nd sem (enrolled)']>20].index
train_df.drop(filter5,inplace=True)
filter6=train_df[train_df['Curricular units 2nd sem (approved)']>19].index
train_df.drop(filter6,inplace=True)
```

```python
train_df.drop(train_df['Course'],inplace=True)
test_df.drop(test_df['Course'],inplace=True)
```

```python
import sklearn

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split

X=train_df.drop(['Target','id'],axis=1)
y=train_df['Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=0)

params={'n_estimators':[100,200,300,400]}

gsearch=GridSearchCV(estimator=RandomForestClassifier(),param_grid=params,scoring='f1_macro')
gsearch.fit(X_train, y_train)
print("The best parameters are:", gsearch.best_params_)
model=gsearch.best_estimator_
print(model.score(X_train,y_train))
```
